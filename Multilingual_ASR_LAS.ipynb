```python
class SpeechDataset(Dataset):
    def __init__(self, df, resample=True):
        self.df = df
        self.specgram = MelSpectrogram() 
        self.resample = resample
    def __len__(self):
        return self.df.shape[0]
    
    def __getitem__(self, idx):
        # preparing audio data
        filename = self.df['audio_path'].iloc[idx]
        waveform, sample_rate = torchaudio.load(filename)
        if sample_rate == 8000:
          waveform = upsample(waveform)
        audio = self.specgram(waveform)
        audio = audio.log2().clamp(min=-50)
        # Normalize input
        audio = (audio - audio.mean())
        audio = audio/audio.abs().max()
        audio = audio.squeeze(0).transpose(0,1)
        # preparing target
        label = self.df['text'].iloc[idx]
        label = [sos] + [char_to_token[i] for i in label] + [eos]
        return audio, torch.tensor(label)
```

```python
# Return utterance, text, utterance length, text length
def collate_train(batch):
    inputs, targets, input_lens, target_lens = [], [], [], []
    for i in range(len(batch)):
      #input audio
      inputs.append(torch.tensor(batch[i][0]))
      #utterance length
      input_lens.append(len(batch[i][0]))
      #target
      targets.append(torch.tensor(batch[i][1][1:]))
      #target length
      target_lens.append(len(batch[i][1])-1)

    #padding
    inputs = pad_sequence(inputs, batch_first=True) 
    targets = pad_sequence(targets, batch_first=True)
    return inputs, targets, torch.tensor(input_lens), torch.tensor(target_lens)
```

# Listener Model

```python
class PBLSTM(nn.Module):
    def __init__(self, in_dim, h_dim):
        super().__init__()

        self.pblstm = nn.LSTM(input_size=in_dim, hidden_size=h_dim, num_layers=1, bidirectional=True, batch_first=True)

    def forward(self, x):
        #unpack the waveform and its length
        x, x_len = pad_packed_sequence(x, batch_first=True)
        x_len = x_len.to(device)
        # Make the time to even length
        x = x[:, :(x.size(1)//2)*2, :]
        # Divide the time to half
        x = x.reshape(x.size(0), x.size(1)//2, x.size(2)*2)
        x_len = x_len // 2 # Divide the time length to reduce it to half
        #pack the waveform and its length again and pass to next layer
        x_packed = pack_padded_sequence(x, lengths=x_len.cpu(), batch_first=True, enforce_sorted=False)
        out, _ = self.pblstm(x_packed)
        return out
```

```python
class Listener(nn.Module):
    def __init__(self, in_dim, h_dim, v_size=128,k_size=128):
        super().__init__()
        #One normal lstm layer
        self.lstm = nn.LSTM(in_dim, h_dim, num_layers=1, bidirectional=True, batch_first=True)
        # 3 pyramidal LSTM layers
        self.pBLSTMs = nn.Sequential(
            PBLSTM(h_dim*4, h_dim),
            PBLSTM(h_dim*4, h_dim),
            PBLSTM(h_dim*4, h_dim)
        )
        #Linear layer for key and value
        self.keyl = nn.Linear(h_dim*2, v_size)
        self.valuel = nn.Linear(h_dim*2, k_size)

    def forward(self, x, x_len):
        x = pack_padded_sequence(x, lengths=x_len.cpu(), batch_first=True, enforce_sorted=False)
        x, _ = self.lstm(x)
        x = self.pBLSTMs(x)

        x, x_len = pad_packed_sequence(x, batch_first=True)
        key = self.keyl(x)
        value = self.valuel(x)
        return key, value, x_len
```

# Attention

```python
class Attention(nn.Module):
    def __init__(self):
        super().__init__()
    def forward(self, d_output, key, value, lens):
        # h_i*s_i
        e = torch.bmm(key, d_output.unsqueeze(2)).squeeze(2)
        mask = torch.arange(key.size(1)).unsqueeze(0) >= lens.unsqueeze(1)
        mask = mask.to(device)
        e.masked_fill_(mask, -1e9)
        attention = nn.functional.softmax(e, dim=1)
        context = torch.bmm(attention.unsqueeze(1), value).squeeze(1)
        return context
```

```python
class AttendAndSpell(nn.Module):
    def __init__(self, vocab_size, d_h_dim, emb_dim, v_size=128, k_size=128):
        super().__init__()
        # create embeddings
        self.embedding = nn.Embedding(vocab_size, emb_dim, padding_idx=0)
        # LSTM CELLS
        self.lstm1 = nn.LSTMCell(emb_dim + v_size, d_h_dim)
        self.lstm2 = nn.LSTMCell(d_h_dim, k_size)
        # Attention layer
        self.attention = Attention()
        # Final Output
        self.char_pred = nn.Linear(k_size + v_size, vocab_size)
        self.v_size = v_size
        self.hidden_dim = d_h_dim

    def forward(self, key, values, encoder_lens, text=None, isTrain=True, teach_force = 0.9, gumbel=False):
        batch_size = key.shape[0]
        if (isTrain == True):
            #use target embedding while training
            max_len =  text.shape[1]
            embeddings = self.embedding(text)
        else:
            #use max length for testing
            max_len = 150
        preds = []
        h_states = [None, None]
        #initial pred is zero
        pred = torch.zeros(batch_size, 1).to(device)
        #initial context
        context = values[:, 0, :]
        for T in range(max_len):
            if (isTrain):
                # Teacher forcing
                teacher_forcing = True if random.random() > teach_force else False
                if teacher_forcing:
                    if T != 0 and gumbel: 
                        #add gumbel noise
                        char_embed = torch.nn.functional.gumbel_softmax(pred).mm(self.embedding.weight)
                    else:
                        char_embed = self.embedding(pred.argmax(dim=-1))
                else:
                    if T == 0:
                        start_char = torch.zeros(batch_size, dtype=torch.long).fill_(char_to_token['<sos>']).to(device)
                        char_embed = self.embedding(start_char)
                    else:
                        # Use ground truth
                        char_embed = embeddings[:, T-1, :]
            else:
                # TESTING
                if T == 0:
                    start_char = torch.zeros(batch_size, dtype=torch.long).fill_(char_to_token['<sos>']).to(device)
                    char_embed = self.embedding(start_char)
                else:
                    char_embed = self.embedding(pred.argmax(dim=-1))

            # concate char_embedding and context and input to decoder
            inp = torch.cat([char_embed, context], dim=1)
            h_states[0] = self.lstm1(inp, h_states[0])

            inp_2 = h_states[0][0]
            h_states[1] = self.lstm2(inp_2, h_states[1])

            #output
            output = h_states[1][0]

            context = self.attention(output, key, values, encoder_lens)
            # concatinate to get prediction
            pred = self.char_pred(torch.cat([output, context], dim=1))
            preds.append(pred.unsqueeze(1))
        return torch.cat(preds, dim=1)
```

# LAS

```python
class LAS(nn.Module):
    def __init__(self, input_dim, vocab_size, listener_hid_dim=256, d_h_dim=512, emb_dim=256, v_size=128, k_size=128):
        super().__init__()
        self.encoder = Listener(input_dim, listener_hid_dim)
        self.decoder = AttendAndSpell(vocab_size, d_h_dim, emb_dim, v_size, k_size)

    def forward(self, speech_input, speech_len, text_input=None, isTrain=True):
        key, value, encoder_lens = self.encoder(speech_input, speech_len)
        if (isTrain == True):
            preds = self.decoder(key, value, encoder_lens, text_input, gumbel=True)
        else:
            preds = self.decoder(key, value, encoder_lens, text=None, isTrain=False)
        return preds
```

# Character Error Rate

```python
def char_error_rate(preds, targets):
    cer = 0.0
    for pred, target in zip(preds, targets):
        dist = Levenshtein.distance(pred, target)
        dist = dist/len(pred)
        cer += dist
    return cer
```

# Word Error Rate

```python
def _levenshtein_distance(ref, hyp):
    m = len(ref)
    n = len(hyp)

    # special case
    if ref == hyp:
        return 0
    if m == 0:
        return n
    if n == 0:
        return m

    if m < n:
        ref, hyp = hyp, ref
        m, n = n, m

    # use O(min(m, n)) space
    distance = np.zeros((2, n + 1), dtype=np.int32)

    # initialize distance matrix
    for j in range(0,n + 1):
        distance[0][j] = j

    # calculate levenshtein distance
    for i in range(1, m + 1):
        prev_row_idx = (i - 1) % 2
        cur_row_idx = i % 2
        distance[cur_row_idx][0] = i
        for j in range(1, n + 1):
            if ref[i - 1] == hyp[j - 1]:
                distance[cur_row_idx][j] = distance[prev_row_idx][j - 1]
            else:
                s_num = distance[prev_row_idx][j - 1] + 1
                i_num = distance[cur_row_idx][j - 1] + 1
                d_num = distance[prev_row_idx][j] + 1
                distance[cur_row_idx][j] = min(s_num, i_num, d_num)

    return distance[m % 2][n]


def word_errors(reference, hypothesis, ignore_case=False, delimiter=' '):
    if ignore_case == True:
        reference = reference.lower()
        hypothesis = hypothesis.lower()

    ref_words = reference.split(delimiter)
    hyp_words = hypothesis.split(delimiter)

    edit_distance = _levenshtein_distance(ref_words, hyp_words)
    return float(edit_distance), len(ref_words)


def WER(reference, hypothesis, ignore_case=False, delimiter=' '):
    wer = 0
    for pred, target in zip(hypothesis, reference):
      edit_distance, ref_len = word_errors(target, pred, ignore_case,
                                          delimiter)

      if ref_len == 0:
          raise ValueError("Reference's word number should be greater than 0.")

      wer = wer + float(edit_distance) / ref_len
    return wer
```

# Prediction to Text

```python
def index_to_letter(batch, stop_tokens):
    sentence_list = []
    for sent in batch:
        curr = ""
        for ch in sent:
            if ch in stop_tokens:
                break
            else:
                curr += token_to_char[ch]
        sentence_list.append(curr)
    return sentence_list
```

# Masking

```python
def generate_mask(lens):
    lens = torch.tensor(lens).to(device)
    max_len = torch.max(lens)

    mask = (torch.arange(0, max_len).repeat(lens.size(0), 1).to(device) < lens.unsqueeze(1).expand(lens.size(0), max_len)).int()
    return mask
```

```python
def train(model, train_loader, criterion, optimizer, epoch, display_batch=50):
    model.train()
    model.to(device)
    batch_loss = 0.0
    train_loss = 0.0
    for batch_idx, (uttr, target, uttr_len, tar_len) in enumerate(tqdm.notebook.tqdm(iter(train_loader))):
        with torch.autograd.set_detect_anomaly(False):
            uttr, target, uttr_len, tar_len = uttr.to(device), target.to(device), uttr_len, tar_len.to(device)
            optimizer.zero_grad()

            #predictions
            preds = model(speech_input=uttr, speech_len=uttr_len, text_input=target, isTrain=True)

            #create mask
            mask = generate_mask(tar_len).to(device)

            #loss
            loss = criterion(preds.view(-1, preds.size(2)), target.view(-1))

            #mask the loss
            m_loss = torch.sum(loss * mask.view(-1)) / torch.sum(mask)
            curr_loss = m_loss.item()
            batch_loss += curr_loss
            train_loss += curr_loss
            m_loss.backward()

            torch.nn.utils.clip_grad_norm(model.parameters(), 2)
            optimizer.step()
            
            if batch_idx % display_batch == (display_batch-1):
                print("Epoch:{}\tBatch:{}\tLoss: {:.5f}".format(epoch, batch_idx, batch_loss/display_batch))
                batch_loss = 0.0
            del uttr
            del target
            del uttr_len
            del tar_len
            torch.cuda.empty_cache()
    print("Final Epoch:{}\tTrainLoss:{:.5f}".format(epoch, train_loss/len(train_loader)))
```

```python
def evaluate(model, test_loader, epoch, display_batch=50):
    model.eval()
    model.to(device)
    numSeq = 0.0
    runningDist = 0.0
    runningWER = 0.0
    for batch_idx, (data, target, uttr_len, tar_len) in enumerate(tqdm.notebook.tqdm(iter(test_loader))):
        data, target, uttr_len, tar_len = data.to(device), target.to(device), uttr_len, tar_len.to(device)
        
        preds = model(speech_input=data, speech_len=uttr_len, text_input=None, isTrain=False)
        predText = index_to_letter(preds.argmax(-1).detach().cpu().numpy(),[char_to_token['<eos>'], char_to_token['<pad>']])
        targetText = index_to_letter(target.detach().cpu().numpy(),[char_to_token['<eos>'], char_to_token['<pad>']])
        cer = char_error_rate(predText, targetText)
        wer = WER(targetText, predText)
        runningDist += cer/len(data)
        numSeq = numSeq + 1
        runningWER += wer/len(data)
            
        del data
        del target
        del uttr_len
        del tar_len
        torch.cuda.empty_cache()
    print("Finished Epoch: {}\tAvg CER: {:.5f}\tAvg WER: {:.5f}".format(epoch, runningDist/numSeq, runningWER/numSeq))
    return runningDist/numSeq
```

# Training

```python
epochs = 40
for epoch in range(epochs):
    # Train
    print("Train \t{} Epoch".format(epoch))
    train(model, train_loader, criterion, optimizer, epoch, display_batch=50)
    # Evaluate
    print("Evaluation \t{} Epoch".format(epoch))
    cer = evaluate(model, test_loader, epoch, display_batch=50)
    scheduler.step(cer)
```

```python
def Sampling(model, test_loader, pred_freq=2):
    model.eval()
    model.to(device)
    numSeq = 0.0
    runningDist = 0.0
    runningWER = 0.0
    for batch_idx, (data, target, uttr_len, tar_len) in enumerate(tqdm.notebook.tqdm(iter(test_loader))):
        data, target, uttr_len, tar_len = data.to(device), target.to(device), uttr_len, tar_len.to(device)
        
        preds = model(speech_input=data, speech_len=uttr_len, batchNum=batch_idx, text_input=None, isTrain=False)
        predText = index_to_letter(preds.argmax(-1).detach().cpu().numpy(),\
                                                 [char_to_token['<eos>'], char_to_token['<pad>']])
        targetText = index_to_letter(target.detach().cpu().numpy(),\
                                               [char_to_token['<eos>'], char_to_token['<pad>']])
        cer = char_error_rate(predText, targetText)
        wer = WER(targetText, predText)
        runningDist += cer/len(data)
        numSeq = numSeq + 1
        runningWER += wer/len(data)

        if batch_idx % pred_freq == (pred_freq-1):
            print("Batch : {}\nPred:\n{}\nTarget:\n{}\n".format(batch_idx, predText[0], targetText[0]))
            
        del data
        del target
        del uttr_len
        del tar_len
        torch.cuda.empty_cache()
    print("Avg CER: {:.5f}\tAvg WER: {:.5f}".format(runningDist/numSeq, runningWER/numSeq))
```

```python
Sampling(model, test_loader)
```
